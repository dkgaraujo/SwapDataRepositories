```python
import os
import pandas as pd
import patoolib
import plotly

from ftplib import FTP
from google.colab import drive

FTP.maxline = 16384 # https://stackoverflow.com/questions/28016444/ftp-library-error-got-more-than-8192-bytes
```

```python
server = 'ftp.cmegroup.com'
ftp = FTP(host=server)

ftp.login()
```

```python
ftp.cwd('sdr/fx')
```

```python
years = ftp.nlst()
```

```python
dirs = []
for year in years:
  dirs += ftp.nlst(year)

print(dirs)
```

```python
files = []
for dir in dirs:
  files += ftp.nlst(dir)
```

```python
GDrivePath = '/content/drive'
drive.mount(GDrivePath)
GDrivePath += '/MyDrive/SwapDataRepositoryData'
```

```python
os.makedirs(GDrivePath, exist_ok=True)
```

```python
for file_orig in files:
  print(file_orig)
  file_name = file_orig.split('/')[-1]
  with open(GDrivePath+'/'+file_name, 'w') as fp:
    res = ftp.retrlines('RETR ' + file_orig, fp.write)
    if not res.startswith('226 Transfer complete'):
      print('Download of file '+file_orig+' failed.')
```

```python
zip_files = [i for i in os.listdir(GDrivePath) if i.split('.')[-1] == 'zip']
```

```python
for z in zip_files:
  print('Extracting '+z+'...')
  patoolib.extract_archive(GDrivePath+'/'+z,  outdir=GDrivePath)
```
