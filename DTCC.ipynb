# DTCC Swap Data

```python
import requests
import io
import os
import zipfile

from datetime import datetime, timedelta
from tqdm import tqdm
```

```python
def create_url(type, year, month, day):
    year = str(year)
    month = str(month)
    day = str(day)
    return f"https://kgc0418-tdw-data-0.s3.amazonaws.com/cftc/eod/CFTC_CUMULATIVE_{type}_{year}_{month}_{day}.zip"
```

```python
types = ['COMMODITIES', 'CREDITS', 'EQUITIES', 'FOREX', 'RATES']
goback = 360 * 4# how many days to go back in time for the data
```

```python
path = 'sdr_data/DTCC/'
```

```python
today = datetime.today()
for i in tqdm(range(goback)):
    date_i = today - timedelta(i)
    for type in types:
        zip_url = create_url(type=type, year=date_i.year, month=date_i.month, day=date_i.day)
        zip_file_name = zip_url.split('/')[-1]
        zip_file_path = os.path.join(path, zip_file_name)
        r = requests.get(zip_url, stream=True, verify=False)
        with open(zip_file_path, 'wb') as fd:
            for chunk in r.iter_content(chunk_size=128):
                fd.write(chunk)
        
```
